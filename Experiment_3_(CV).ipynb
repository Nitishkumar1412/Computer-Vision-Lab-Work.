#**AIM: To Implement smoothing or averaging filter in spatial domain.**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('bmw.jpeg')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

#**Kernel Size = (5,5)**

```python
kernel_size = (5, 5)
smoothed_image = cv2.blur(image, kernel_size)
```

```python
smoothed_image_rgb = cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.title("Smoothed Image (5x5 Average Filter)")
plt.imshow(smoothed_image_rgb)
plt.axis('off')
```

```python
plt.tight_layout()
plt.show()
```
