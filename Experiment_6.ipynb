#**AIM: Edge Detection using Non-Linear Filtering: Canny Edge Detector**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Lawrance.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
filtered_image = cv2.medianBlur(image, ksize=5)
```

```python
edges = cv2.Canny(filtered_image, threshold1=50, threshold2=150)
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.title('Edge Detection using Non-Linear Filtering: Canny Edge Detector')
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.show()
```
