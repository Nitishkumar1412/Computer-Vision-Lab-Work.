#**AIM: Edge detection on color image**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Lion.jpeg')
```

```python
blue_channel, green_channel, red_channel = cv2.split(image)
```

```python
edges_blue = cv2.Canny(blue_channel, threshold1=50, threshold2=150)
edges_green = cv2.Canny(green_channel, threshold1=50, threshold2=150)
edges_red = cv2.Canny(red_channel, threshold1=50, threshold2=150)
```

```python
combined_edges = cv2.bitwise_or( cv2.bitwise_or(edges_blue, edges_green),edges_red)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Edge Detection on Color Image')
plt.imshow(combined_edges, cmap='gray')

plt.show()
```
