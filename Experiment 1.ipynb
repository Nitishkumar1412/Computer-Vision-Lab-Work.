**Importing libraries**

```python
from skimage import io, color
import matplotlib.pyplot as plt
```

**Load the image**

```python
image = io.imread('virat.jpeg')
```

Convert to greyscale

```python
gray_image = color.rgb2gray(image)
```

**Display the image**

```python
plt.imshow(image)
plt.title("Loaded image")
plt.axis('off')
plt.show()
```

**Display the grey scale image**

```python
plt.imshow(gray_image, cmap='gray')
plt.title("Grey scale image")
plt.axis('off')
plt.show()
```

**Display the viridis scale image**

```python
plt.imshow(gray_image, cmap='viridis')
plt.title("Grey scale image")
plt.axis('off')
plt.show()
```

**Display the jet scale image**

```python
plt.imshow(gray_image, cmap='jet')
plt.title("Grey scale image")
plt.axis('off')
plt.show()
```

**Display the hot scale image**

```python
plt.imshow(gray_image, cmap='hot')
plt.title("Grey scale image")
plt.axis('off')
plt.show()
```

**Display the cool scale image**

```python
plt.imshow(gray_image, cmap='cool')
plt.title("Grey scale image")
plt.axis('off')
plt.show()
```
