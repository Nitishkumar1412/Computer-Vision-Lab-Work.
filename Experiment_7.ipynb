#**Aim: Edge Detection using Non-Linear Filtering: Laplacian Edge Detection**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Dodge.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
filtered_image = cv2.medianBlur(image, ksize=5)
```

```python
blurred_image = cv2.GaussianBlur(filtered_image, (5, 5), 0)
```

```python
laplacian_edges = cv2.Laplacian(blurred_image, cv2.CV_64F)
```

```python
abs_laplacian_edges = np.abs(laplacian_edges)
```

```python
scaled_laplacian_edges = np.uint8(255 * abs_laplacian_edges / np.max(abs_laplacian_edges))
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Laplacian Edge Detection')
plt.imshow(scaled_laplacian_edges, cmap='gray')

plt.show()
```
